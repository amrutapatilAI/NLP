**Ans 1.c)** A model of polynomial of degree 5 is appropriate for the data.The temperature at 2020 is predicted as 60 °$F$

```r
plot(temp~year, data = aatemp)
abline(v = 1930, lty = 5)
lhs <- function(x)ifelse(x<=1930,mean(aatemp$temp),0)
rhs <- function(x)ifelse(x>1930,x-1930,0)
mdl4 <- lm(temp~ lhs(year) + rhs(year), data = aatemp)
x <- seq(1854,2000, by=1)
py <-mdl4$coef[1] + mdl4$coef[2]*lhs(x) + mdl4$coef[3]*rhs(x)
lines(x,py,lty = 2)
```

```r
summary(mdl4)
```

**Ans 1.d)** There is a weak negative linear relationship b/w "year" and "temp" after 1930. The model seems to have some error.

```r
attach(aatemp)
knots <- c(1850,1850,1850,1850,1868.75,1906.25,1943.75,1981.25,2000,2000,2000,2000)
byear <- splineDesign(knots,year)
mdl5 <- lm(temp~byear-1)
matplot(year,byear,type = "l", col = 1)
matplot(year, cbind(temp,mdl5$fit), type = "pl", ylab = "year", pch = 20, lty = 1, col = 1)
```

**Ans 1.e)** From the plot it is seen that the cubic spline fit is not better than linear fit.

```r
data(ozone)
```

```r
mdl10 <- lm(O3~temp + humidity + ibh ,ozone)
summary(mdl10)
```

```r
library(MASS)
## Box-Cox method for Savings data
boxcox(mdl10, plotit=T, lambda=seq(0.1,0.4,by=0.05))
```

Based on the graph of log-likelihood $λ$ above, the maximum likelihood is at $\lambda \approx 0.28$. To easily interpret, we will set $\lambda$ at $0.25$.

```r
mdl11 <- lm(O3^(1/4)~temp + humidity + ibh ,ozone)
summary(mdl11)
```

**Ans 2**

As we can see the summary above, the p-values from all predictors are extremely low, meaning that the model fits very well.
